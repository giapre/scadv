```python
import numpy as np
import pandas as pd
from synth_pat.paths import Paths
```

```python
# Baselines per receptor
baseline_vec = np.array([0.5, 1.0, 0.25])  # D1, D2, S

med_dic = {
    'clozapine': {'Z_D1': 64/10000, 'Z_D2': 38/10000, 'Z_S': 1870/10000},
    'aripiprazole': {'Z_D1': 5/10000, 'Z_D2': 303/10000, 'Z_S': 1150/10000},
    'amisulpride': {'Z_D1': 1/10000, 'Z_D2': 3333/10000, 'Z_S': 1.2/10000},
    'olanzapine': {'Z_D1': 476/10000, 'Z_D2': 22727/10000, 'Z_S': 25641/10000}
}

k_vals = np.linspace(0, 0.5, 20)
k_max = k_vals.max()

rows = []

for med, ratios in med_dic.items():

    vec = np.array([
        ratios['Z_D1'],
        ratios['Z_D2'],
        ratios['Z_S']
    ])

    # normalize receptor profile
    vec_norm = vec / vec.max()

    for ik, k in enumerate(k_vals):

        # scale so that k_max → zero
        scaled_effect = (k / k_max) * baseline_vec * vec_norm

        Z = baseline_vec - scaled_effect

        rows.append({
            "medication": med,
            "k": k,
            "k_i": ik,
            "Z_D1": Z[0],
            "Z_D2": Z[1],
            "Z_S": Z[2]
        })

df = pd.DataFrame(rows)
df.to_csv(f'{Paths.RESOURCES}/mediction_sweep_params.csv', index=False)
```
