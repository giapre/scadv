```python
import os
import numpy as np
from synth_pat.paths import Paths

# reshape snakemake results so that you can run 4_extract_synth_datafeatures.py

res_dir = '/home/prior/gp_home/scadv/synth_pat/scripts/9bis_huifang_ppc/results'

for pid in os.listdir(res_dir):
    pid_dir = f'{res_dir}/{pid}/bold_sweep'

    if not os.path.isdir(pid_dir):
        continue

    bolds = []
    para = []
    files = sorted(os.listdir(pid_dir))  # keep order consistent

    for sim in files:
        file_path = f'{pid_dir}/{sim}'
        a = np.load(file_path)
        bolds.append(a['bold'])
        para.append(a['params'])

    bolds = np.array(bolds)
    para = np.array(para)

    # reshape if needed
    bolds = np.squeeze(bolds, axis=-1)      # (n runs, time, regions)
    bolds = np.transpose(bolds, (1, 2, 0)) # (time, regions, n runs)

    save_dir = f'{Paths.RESULTS}/{pid}'
    os.makedirs(save_dir, exist_ok=True)

    save_file = f'{save_dir}/{Paths.TYPE_OF_SWEEP}.npz'
    if os.path.exists(save_file):
        print(f'{save_file} already exists, going to next subject')
        continue
    
    np.savez(save_file, bold=bolds, params=para, param_names=a['param_names'])
    print(f"Saved {save_file}, now deleting individual runs...")

    for sim in files:
        file_path = f'{pid_dir}/{sim}'
        os.remove(file_path)

    print(f"Cleaned up {len(files)} files for {pid}")
```
