```python
import pandas as pd
import numpy as np
from synth_pat.paths import Paths
```

```python
for pid in ['sub-2015052501', 'sub-2015060902', 'sub-2019012101', 'sub-2015060401', 'sub-2015061901', 'sub-2019012402']:
    pid_idx = int(pid.split('sub-')[1])
    emp_data = pd.read_csv(f'{Paths.DATA}/ALL_daniela_full_extracted_features.csv', index_col = 'pid')
    synth_data = pd.read_csv(f'{Paths.RESULTS}/{pid}/{Paths.TYPE_OF_SWEEP}_extracted_features.csv', index_col = 0)
    print(f'Comparing {Paths.TYPE_OF_SWEEP}_extracted_features.csv for {pid}')
    params = ['ws', 'njdopa_ctx', 'njdopa_str']
    feat_cols = [col for col in synth_data.columns if col not in params]
    emp_data_pid = emp_data.loc[pid_idx, feat_cols]

    err_line = []
    for i in synth_data.index:
        line_feat = synth_data.loc[i,feat_cols]
        err_line.append(np.sum(np.abs(line_feat - emp_data_pid)))

    err_line = np.array(err_line)
    best_idx = np.argsort(err_line)[:100]
    best_params = synth_data.loc[best_idx, params]
    #best_params.to_csv(f'{Paths.DERIVATIVES}/{pid}/best_ppc_params_for_medication.csv', index=False)
```
