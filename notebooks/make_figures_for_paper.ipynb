```python
def basic_3d_plot(sweep_df, p1_name, p2_name, p3_name, var_to_plot, var_to_plot_title, ax=None):

    import numpy as np
    import matplotlib.pyplot as plt
    from scipy.interpolate import griddata
    cmap = 'YlGnBu'
    # ---------------------------------------
    # Extract data
    # ---------------------------------------
    x = np.log10(sweep_df[p1_name].astype(float).values)
    y = np.log10(sweep_df[p2_name].astype(float).values)
    z = np.log10(sweep_df[p3_name].astype(float).values)
    c = sweep_df[var_to_plot].astype(float).values

    # ---------------------------------------
    # Point sizes scaled by variable
    # ---------------------------------------
    if c.max() == c.min():
        sizes = np.ones_like(c) * 10
    else:
        sizes = 1 + (5*(c - c.min()) / (c.max() - c.min()))**3

    # ---------------------------------------
    # Create figure if needed
    # ---------------------------------------
    if ax is None:
        fig = plt.figure(figsize=(9,7))
        ax = fig.add_subplot(111, projection='3d')
    else:
        fig = ax.figure

    sc = ax.scatter(x, y, z,
                    c=c,
                    cmap=cmap,
                    s=sizes,
                    alpha=0.9,
                    edgecolors='none')

    # ---------------------------------------
    # Grid for interpolation
    # ---------------------------------------
    xi = np.linspace(x.min(), x.max(), 40)
    yi = np.linspace(y.min(), y.max(), 40)
    zi = np.linspace(z.min(), z.max(), 40)

    XI, YI = np.meshgrid(xi, yi)
    XI2, ZI = np.meshgrid(xi, zi)
    YI2, ZI2 = np.meshgrid(yi, zi)

    # ---------------------------------------
    # Interpolate for contour planes
    # ---------------------------------------
    C_xy = griddata((x, y), c, (XI, YI), method='linear')
    C_xz = griddata((x, z), c, (XI2, ZI), method='linear')
    C_yz = griddata((y, z), c, (YI2, ZI2), method='linear')

    # ---------------------------------------
    # Add padding around data
    # ---------------------------------------
    pad = 0.15

    xr = x.max() - x.min()
    yr = y.max() - y.min()
    zr = z.max() - z.min()

    xmin = x.min() - pad * xr
    xmax = x.max() + pad * xr
    ymin = y.min() - pad * yr
    ymax = y.max() + pad * yr
    zmin = z.min() - pad * zr
    zmax = z.max() + pad * zr

    ax.set_xlim(xmin, xmax)
    ax.set_ylim(ymin, ymax)
    ax.set_zlim(zmin, zmax)

    # ---------------------------------------
    # Planes aligned with cube walls
    # ---------------------------------------
    z_plane = zmin   # bottom XY plane
    y_plane = ymax   # back XZ plane
    x_plane = xmin   # left YZ plane

    # ---------------------------------------
    # Filled contour heatmaps (transparent)
    # ---------------------------------------
    ax.contourf(XI, YI, C_xy,
                zdir='z',
                offset=z_plane,
                cmap=cmap,
                levels=15,
                alpha=0.3)

    ax.contourf(XI2, C_xz, ZI,
                zdir='y',
                offset=y_plane,
                cmap=cmap,
                levels=15,
                alpha=0.3)

    ax.contourf(C_yz, YI2, ZI2,
                zdir='x',
                offset=x_plane,
                cmap=cmap,
                levels=15,
                alpha=0.3)

    # ---------------------------------------
    # Labels and title
    # ---------------------------------------
    ax.set_xlabel(p1_name)
    ax.set_ylabel(p2_name)
    ax.set_zlabel(p3_name)

    ax.set_title(f'3D Scatter of {var_to_plot_title}')

    # Optional cleaner look
    ax.grid(False)
    cbar = fig.colorbar(sc, shrink=0.6, pad=0.1,  label=var_to_plot)

    cbar.set_label("")  # empty string removes label

    cbar.outline.set_visible(False)

    cbar.ax.tick_params(width=0.5, labelsize=10)

    plt.show()

    return sc
```

```python
p1_name = 'ws'
p2_name = 'njdopa_ctx'
p3_name = 'njdopa_str'
params = [p1_name, p2_name, p3_name]
var_to_plot = 'L.PU_ALFF'
var_to_plot_title = 'ALFF of Left Putamen'
basic_3d_plot(sweep_df, p1_name, p2_name, p3_name, var_to_plot, var_to_plot_title, ax=None)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from statsmodels.nonparametric.smoothers_lowess import lowess

vars_x = ['ws', 'njdopa_ctx', 'njdopa_str']
y = 'VAR_FCD'

Y_data = sweep_df[y]
ymin, ymax = Y_data.min()-0.5*Y_data.min(), Y_data.max()+0.1*Y_data.max()

fig = plt.figure(figsize=(14,5))

# main layout
outer = fig.add_gridspec(1, 3, width_ratios=[1.2, 10, 0.3], wspace=0.2)

# KDE axis
ax_kde = fig.add_subplot(outer[0])

# scatter sub-grid (very small spacing)
scatter_grid = outer[1].subgridspec(1, 3, wspace=0.05)

scatter_axes = []

# --- KDE ---
sns.kdeplot(
    y=Y_data,
    fill=True,
    linewidth=0,
    ax=ax_kde
)

ax_kde.set_ylim(ymin, ymax)
ax_kde.set_xlabel("density")
ax_kde.set_ylabel("Fluidity")
ax_kde.invert_xaxis()

ax_kde.yaxis.tick_right()
ax_kde.yaxis.set_label_position("left")

for spine in ax_kde.spines.values():
    spine.set_visible(False)

# --- scatter plots ---
for i, x in enumerate(vars_x):

    ax = fig.add_subplot(scatter_grid[i], sharey=ax_kde)
    scatter_axes.append(ax)

    X = sweep_df[x]
    Y = sweep_df[y]

    sc = ax.scatter(
        X, Y,
        c=Y,
        cmap='viridis',
        alpha=0.25,
        vmin=ymin,
        vmax=ymax,
    )

    smoothed = lowess(Y, X, frac=0.3)
    ax.plot(smoothed[:,0], smoothed[:,1], '--', color='gray', linewidth=2)

    ax.set_xscale('log')
    ax.set_xlabel(x)
    ax.set_ylim(ymin, ymax)

    ax.tick_params(labelleft=False)
    ax.tick_params(axis='both', length=0)

    for spine in ax.spines.values():
        spine.set_color('gray')

# colorbar
cax = fig.add_subplot(outer[2])
cbar = fig.colorbar(sc, cax=cax)
cbar.outline.set_visible(False)

plt.show()
```
