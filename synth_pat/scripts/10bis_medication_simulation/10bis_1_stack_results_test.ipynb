```python
import os
import numpy as np
from synth_pat.paths import Paths

# reshape snakemake results so that you can run 4_extract_synth_datafeatures.py

res_dir = '/home/prior/gp_home/scadv/synth_pat/scripts/10bis_medication_simulation/results'

for pid in os.listdir(res_dir):
    pid_dir = f'{res_dir}/{pid}/'

    if not os.path.isdir(pid_dir):
        print(f'Skipping')
        continue

    save_dir = f'{Paths.RESULTS}/{pid}'
    os.makedirs(save_dir, exist_ok=True)
    save_file = f'{save_dir}/{Paths.TYPE_OF_SWEEP}_medication.npz'
    if os.path.exists(save_file):
        print(f'{save_file} already exists, going to next subject')
        continue

    bolds = []
    med_params = []
    medication_name = []
    est_params = []
    med_zi = []
    files = sorted(os.listdir(pid_dir))  # keep order consistent

    for sim in files:
        med_zi.append(int(sim.split('_')[1])) # taking the index used for finding the medication values in the medication csv 
        file_path = f'{pid_dir}/{sim}'
        a = np.load(file_path)
        bolds.append(a['bold'])
        med_params.append(a['med_params'])
        medication_name.append(a['medication_name'])
        est_params.append(a['est_params'])
    med_param_names = a['med_param_names']
    est_param_names = a['est_param_names']

    bolds = np.array(bolds)
    med_params = np.array(med_params)
    medication_name = np.array(medication_name)
    est_params = np.array(est_params)

    bolds = np.squeeze(bolds, axis=-1)      # (n runs, time, regions)
    bolds = np.transpose(bolds, (1, 2, 0)) # (time, regions, n runs)
    
    np.savez(save_file, bold=bolds, med_zi=med_zi, med_params=med_params, med_param_names=med_param_names, medication_name=medication_name, est_params=est_params, est_param_names=est_param_names)
    print(f"Saved {save_file}, now deleting individual runs...")

    for sim in files:
        file_path = f'{pid_dir}/{sim}'
        os.remove(file_path)

    print(f"Cleaned up {len(files)} files for {pid}")

```
